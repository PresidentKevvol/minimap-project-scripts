```python
import json
```

```python
dataset_location = '../floor_data'
dataset_files = [
    'signal_floor_B_2022-11-29.json',
    'signal_floor_1_2022-11-29.json',
    'signal_floor_2_2022-11-29.json',
    'signal_floor_3_2022-11-29.json',
]

data_json = []
```

```python
floors = ['B', '1', '2', '3']
```

```python
for i in range(len(dataset_files)):
    f = open(dataset_location + '/' + dataset_files[i])
    d = json.loads(f.read())
    data_json.append(d)
```

```python
bssid_set = {}
names = {}

for c in data_json:
    for p in c:
        for pt in p['signals']:
            bssid = pt['bssid']
            ssid = pt['ssid']
            
            if bssid not in names:
                names[bssid] = ssid
            if bssid not in bssid_set:
                bssid_set[bssid] = 0
            bssid_set[bssid] += 1
```

```python
data_json[1][45]
```

```python
names
```

```python
key_aps = ['AirYorkGUEST', 'AirYorkPLUS', 'WiFi-Info', 'eduroam']
```

```python
bssid_set_filtered = {}
names_filtered = {}

for bssid in names:
    if names[bssid] in key_aps:
        names_filtered[bssid] = names[bssid]
        bssid_set_filtered[bssid] = bssid_set[bssid]
```

```python
len(bssid_set_filtered)
```

```python
data_columns = names_filtered.keys()
```

```python
import pandas as pd
```

```python
dataset_size = sum([len(c) for c in data_json])
```

```python
pd_frame = {'floor': ['', ] * dataset_size}
for bssid in names_filtered:
    pd_frame[bssid] = [float('NaN'), ] * dataset_size
```

```python
ct = 0

nonnull = 0

for i in range(len(data_json)):
    dclass = data_json[i]
    cur_floor = floors[i]
    for point in dclass:
        # record floor
        pd_frame['floor'][ct] = cur_floor
        # each access point in this sample point
        for ap in point['signals']:
            bssid = ap['bssid']
            ssid = ap['ssid']
            if ssid in key_aps:
                pd_frame[bssid][ct] = ap['quality']
                nonnull += 1
        ct += 1

df = pd.DataFrame(data=pd_frame)
```
