```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
class WallSegment(object):
    def __init__(self, end1: (float, float), end2: (float, float), damp_factor: float):
        self.end1 = end1
        self.end2 = end2
        self.damp_factor = damp_factor
    
    def pyplot_line(self, ax, style='-', linewidth=2, color='blue'):
        ax.plot([self.end1[0], self.end2[0]], [self.end1[1], self.end2[1]], style, linewidth=linewidth, color=color)
```

```python
walls = [
    WallSegment((-4.6, 6.0), (3.4, 6.0), 5.0),
    WallSegment((3.4, 6.0), (3.4, -2.8), 5.0),
    WallSegment((1.2, 3.0), (1.2, 6.0), 2.0),
    WallSegment((-0.4, 2.2), (3.4, 2.2), 2.0),
    WallSegment((-0.4, 0.8), (3.4, 0.8), 2.0),
    WallSegment((-0.4, -0.6), (3.4, -0.6), 2.0),
]
```

```python
fig, ax = plt.subplots(figsize=(15, 15), dpi=80)
ax.set_aspect('equal', adjustable='box')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

[w.pyplot_line(ax) for w in walls]
```
